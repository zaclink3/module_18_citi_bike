```python
import pandas as pd
import glob

# create a list of all CSV files in a folder
path = '*.csv'
files = glob.glob(path)

# create an empty list to store the data frames
dfs = []

# loop through each CSV file and append it to the list of data frames
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

# concatenate all data frames in the list into a single data frame
stacked_df = pd.concat(dfs)

```

```python
df = stacked_df.drop(['starttime','stoptime','start station id','start station name','start station latitude','start station longitude','end station id','end station name','end station latitude','end station longitude'], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={'start_lat':'start_latitude','start_lng':'start_longitude','end_lat':'end_latitude', 'end_lng':'end_longitude'})
```
